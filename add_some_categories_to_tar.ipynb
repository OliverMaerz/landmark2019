```python
import pandas as pd
import os
import tarfile
```

```python
csv_path = os.path.join(os.sep, "mnt", "lm", "csv", "train.csv")
csv_path
```

```python
img_path = os.path.join(os.sep, "mnt", "lm", "img-min300-categ")
img_path
```

```python
tar_path = os.path.join(os.sep, "mnt", "bucket", "img-min300-cat-5000.tar")
tar_path
```

```python
labels_df = pd.read_csv(csv_path) 
labels_df
```

```python
# get a series with the category (landmark_id) and the count of images in that category 
landmark_df = labels_df.groupby("landmark_id").size()
```

```python
# sort descending by number of images
landmark_df.sort_values(ascending=False, inplace=True)
```

```python
# Get the first 5000 landmark_id's but skip the first two (ETH, Hawaii)
landmark_df[2:5002]
```

```python
# Create tar file to add the images to
with tarfile.open(name=tar_path, mode="w", bufsize=10*1024*1024) as tar_file_handle: 
    # Skip the first two categories (eth zurich image collection and hawaii plants) 
    # and loop over 1000 landmark_id with the most images
    for landmark_id, img_count in landmark_df[2:1001].items():
        for root, dirs, files in os.walk(os.path.join(img_path, str(landmark_id))):
            print("Adding images from landmark_id/category: {}".format(landmark_id))
            for file in files:
                # Now add the image
                tar_file_handle.add(os.path.join(root, file))
                                         
```
