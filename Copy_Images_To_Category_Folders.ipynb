### Notebook to copy images to folders for each category

```python
import pandas as pd
import os
# import sys
from shutil import copyfile
```

```python
csv_path = os.path.join(os.sep, "mnt", "lm", "csv", "train.csv")
csv_path
```

```python
orig_img_path = os.path.join(os.sep, "mnt", "lm", "images-max300")
orig_img_path
```

```python
new_img_path = os.path.join(os.sep, "mnt", "lm", "img-max300-categ")
new_img_path
```

```python
labels_df = pd.read_csv(csv_path) 
labels_df
```

```python
# make the id column the index
labels_df.set_index("id", inplace=True)
labels_df
```

```python
# find the label (=landmark_id) for a given filename 
def get_label(fname):
  return labels_df.loc[os.path.splitext(os.path.basename(fname))[0]]["landmark_id"]
```

```python
# now loop through all images
for path, sub_dirs, file_names in os.walk(orig_img_path):
    print("Opening folder: "+path)
    for file_name in file_names:
        base_filename, file_extension = os.path.splitext(file_name)
        if file_extension.lower() == '.jpg':
            # found a jpg now lookup the category and copy it there
            # base_filename is the filename without extension - so the id 
            label = get_label(base_filename)
            label_dir = os.path.join(new_img_path, str(label))
            if not os.path.exists(label_dir):
                os.makedirs(label_dir)
            copyfile(os.path.join(path, file_name), os.path.join(label_dir, file_name))
        else:
            print("found a non-jpg with extension: "+file_extension.lower())
```
